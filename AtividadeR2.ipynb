```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("/casos covid 2021 2022.xlsx")
```

```python
x = df.iloc[:,:-1].values
```

```python
y = df.iloc[:, -1].values
```

```python
plt.scatter(x,y, color = 'red')
plt.title('Casos de Covid - 2021 e 2022')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(x,y)
LinearRegression()
ypred = regressor.predict(x)
plt.scatter(x,y, color = 'purple')
plt.title("Casos de Covid - 2021 e 2022")
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x, ypred, color = 'black')
```

```python
regressor.score(x,y)
```

```python
regressor.predict([[0.3]])
```

```python
regressor.coef_
```

```python
regressor.intercept_
```
